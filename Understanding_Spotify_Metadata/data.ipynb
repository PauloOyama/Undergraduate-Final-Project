```python
import pandas as pd
```

```python
a = pd.read_csv("metadados_dataset.csv")
```

```python
a.head()
```

```python
a.columns
```

```python
metadata = ["danceability","key","loudness","mode","speechiness","acousticness","instrumentalness","liveness","valence","tempo","duration_ms","time_signature"]
```

```python
df = a[metadata]
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt

# plot the data
fig = plt.figure()
ax = fig.add_subplot(111)
# plot x,y data with c as the color vector, set the line width of the markers to 0
scatter = ax.scatter(df["loudness"], df["acousticness"], lw=0)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="lower left", title="Classes")
ax.add_artist(legend1)

# ax.legend()
# ax.grid(True)
plt.xlabel("Instrumental")
plt.ylabel("Durao")
plt.show()
```
